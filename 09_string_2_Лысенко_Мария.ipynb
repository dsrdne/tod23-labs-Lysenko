Считайте, что в слове есть опечатка, если данное слово не содержится в списке `words`. 

```python
text = '''с велечайшим усилием выбравшись из потока убегающих людей Кутузов со свитой уменьшевшейся вдвое поехал на звуки выстрелов русских орудий'''
```

### Расстояние редактирования

1.1 Загрузите предобработанные описания рецептов из файла `preprocessed_descriptions.csv`. Получите набор уникальных слов `words`, содержащихся в текстах описаний рецептов (воспользуйтесь `word_tokenize` из `nltk`). 

```python
with open('preprocessed_descriptions.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    wordsinprep = set()
    for row in reader:
        description = row[2]
        tokens = word_tokenize(description)
        wordsinprep.update(tokens)

print(wordsinprep)
```

1.2 Сгенерируйте 5 пар случайно выбранных слов и посчитайте между ними расстояние редактирования.

```python
word_pairs = random.sample(wordsinprep, k=10)
for i in range(0, len(word_pairs), 2):
    word1 = word_pairs[i]
    word2 = word_pairs[i+1]
    
    distance = edit_distance(word1, word2)
    print(f"Расстояние Левенштейна между '{word1}' и '{word2}': {distance}")
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
def find_closest_words(word, wordsinprep, k):
    distances = [(w, edit_distance(word, w)) for w in wordsinprep]
    sorted_distances = sorted(distances, key=lambda x: x[1])
    closest_words = [w[0] for w in sorted_distances[:k]]
    return closest_words
```

```python
word = 'hello'
k = 3
```

```python
closest_words = find_closest_words(word, wordsinprep, k)
print(f"Ближайшие {k} слов(а) к '{word}': {closest_words}")
```

3.2 Вычислите близость между каждой парой рецептов, выбранных в задании 3.1, используя косинусное расстояние (`scipy.spatial.distance.cosine`) Результаты оформите в виде таблицы `pd.DataFrame`. В качестве названий строк и столбцов используйте названия рецептов.

```python
from scipy.spatial.distance import cosine

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(df['preprocessed_descriptions']).toarray()
df_1 = pd.DataFrame(index = df['name'], columns = df['name'])
for i, r1 in enumerate(df['name']):
    for j, r2 in enumerate(df['name']):
        df_1.at[r1, r2] = 1 - cosine(vectors[i], vectors[j])
df_1 = df_1.reset_index().rename(columns={'name': ' '})
df_1 = df_1.set_index(' ')
df_1
```
